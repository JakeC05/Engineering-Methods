```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = fig.add_subplot(projection='3d')

FAAB = np.array([38,100,26,344,268,195,600,618,698]) #FAAB

WASTE = np.array([0,40,152,13,173,251,42,112,234]) #WASTE

NET_WORTH = np.array([962,860,822,643,559,554,358,270,68]) #Net Worth

ax.scatter(FAAB, WASTE, NET_WORTH)
ax.set_xlabel('FAAB')
ax.set_ylabel('WASTE')
ax.set_zlabel('Net Worth')

plt.show()
```

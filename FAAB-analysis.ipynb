```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = fig.add_subplot(projection='3d')

FAAB = np.array([25,75,266,48,94,384,664,167,243]) #FAAB

WASTE = np.array([152,629,220,179,40,47,234,112,26]) #WASTE

NET_WORTH = np.array([823,296,514,773,866,569,102,721,731]) #Net Worth

ax.scatter(FAAB, WASTE, NET_WORTH)
ax.set_xlabel('FAAB')
ax.set_ylabel('WASTE')
ax.set_zlabel('Net Worth')

plt.show()
```

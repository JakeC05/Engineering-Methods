### a. Root Finding
Use Newton-Raphson method to find root of f(x) = x^3 - x - 2 with the initial guess x₀ = 1.5 and tolerance = 1e-5.

```python
#jake was here
import numpy as np

def f(x):
    return x**3 -x -2

x0=1.5
tol=1e-30

for i in range(1000):
    x1 = x0 - f(x0)/(3*x0**2 -1)
    if abs(x1 - x0) < tol:
        break
    x0 = x1
print(f"Root found: {round(x1,4)} in {i+1} iterations")
```

Use central difference to estimate the first and second derivatives of y =  1/(x^2 + 3x + 5) at x = 5.

```python
# Your code here
import numpy as np
def f(x):
    return 1/(x**2+3*x+5)

x0=5
h=0.0001

def df(x):
    return (f(x+h)-f(x-h))/(2*h)

print(df(x0))

def d2f(x):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)

print(round(d2f(x0),3))
```

### c. Numerical Integration
Use Simpson's Rule to approximate the integral of 1 / (1 + x^2) from 0 to 2.

```python
# Your code here
import numpy as np
a,b=0,2
n=10
x=np.linspace(a,b,n+1)
def f(x):
    return 1/(1+x**2)
def simpson_rule(f,a,b,n):
    if n %2 !=0:
        raise ValueError("n must be even")
    h=(b-a)/n
    result=f(a)+f(b)
    for i in range(1,n):
        coeff=4 if i%2 !=0 else 2
        
        result+=coeff*f(a+i*h)
    return result*h/3
simpson_int=simpson_rule(f,a,b,n)

display(round(simpson_int,6))
```

### e. ODE BVP
Using FDM to solve d²y/dx² = -π²y with boundary conditions y(0)=0, y(1)=0.

Algorithm: 
1. Discretization
2. Using central difference to approximate the second derivative.
3. Formulate the linear system of equations Ax= b

Hint: $$ y_{i-1} + (-2 - \pi^2h^2)y_i + y_{i+1} = 0 $$
4. Incorporate boundary conditions
5. Solve using NumPy's linear solver

```python
# Your code here
import numpy as np

n = 10                
h = 1 / (n + 1)       
x = np.linspace(0, 1, n + 2)

A = np.zeros((n, n))
main_diag_value = -2 - (np.pi**2) * h**2

for i in range(n):
    A[i, i] = main_diag_value
    if i > 0:
        A[i, i-1] = 1
    if i < n-1:
        A[i, i+1] = 1

b = np.zeros(n)

y_inner = np.linalg.solve(A, b)

y = np.concatenate(([0], y_inner, [0]))

y
```
